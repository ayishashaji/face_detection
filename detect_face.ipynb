```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r'C:\DL projects\face_detection\pngtree-group-of-girls-posing-for-a-picture-picture-image_2781639.jpg')
img
```

```python
cv2.imshow('IMAGE',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
# creating an object for classifying a image
face_cascade=cv2.CascadeClassifier(r'C:\DL projects\face_detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_grey=cv2.cvtColor(conv_img,cv2.COLOR_RGB2GRAY)
img_grey
```

```python
face=face_cascade.detectMultiScale(img_grey,minNeighbors=4)
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,0,255),thickness=4)
```

```python
plt.imshow(conv_img)
```

FACE DETECTION FROM VIDEO.

```python
video=cv2.VideoCapture(r'C:\DL projects\face_detection\video.mp4')
while True:
    success,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=10)
    # print(faces)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,122,232),3)
    cv2.imshow('VIDEO',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()        
```

FACE DETECTION FROM WEBCAM

```python
video = cv2.VideoCapture(0)
while True:
    success,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,minNeighbors=10)
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,122,232),3)
    cv2.imshow('WEBCAM',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()    
```
